```python
#Creo las nuevas columnas "mention" y "hashtag" que contienen un 1 si hay una arroba y 0 sino, 
#y 1 si hay numeral y 0 sino, respectivamente
twit["mention"] = twit["text"].apply(lambda x: 1 if "@" in x else 0)
twit["hashtag"] = twit["text"].apply(lambda x: 1 if "#" in x else 0)
```

```python
#Agrego como columna de interes el largo del texto
twit["text_long"] = twit["text"].apply(lambda x: len(x))
```

```python
twit.head()
```

```python
twit["text_long"].describe()
```

```python
#Grafico Histograma de la distribucion de longitud de los twits.
fig = plt.figure()
ax = fig.add_subplot(111)
plt.title("Longitud del texto, Histograma", fontsize=20,fontweight="bold")
ax.set_xlabel("Cantidad de caracteres en el twit")

ax.axvline(twit.text_long.mean(), color = "b", linestyle = "--")
plt.text(twit.text_long.mean()+1,800,'Valor Promedio',rotation=90, color = "b")

ax.axvline(twit.text_long.quantile(0.5), color = "r", linestyle = "--")
plt.text(twit.text_long.quantile(.5)+1,800,'segundo cuartil',rotation=90, color = "r")

ax.axvline(twit.text_long.quantile(0.25), color = "black", linestyle = "--")
plt.text(twit.text_long.quantile(.25)+1,800,'primer cuartil',rotation=90, color = "black")

ax.axvline(twit.text_long.quantile(0.75), color = "salmon", linestyle = "--")
plt.text(twit.text_long.quantile(.75)+1,800,'tercer cuartil',rotation=90, color = "salmon")

twit["text_long"].plot.hist(color="green", figsize=(12,4),bins=20)
plt.show()
```

```python
#Hago un analisis segun keyword
twit.keyword.describe()
```

```python
twit.keyword.fillna('sinKeyword',inplace=True)
twit.head(10)
```

```python
#Calculo el porcentaje de tweets verdaderos sin keyword
prcSinKeyword = (twit[twit['keyword']=='sinKeyword'].target.sum())*100/(twit[twit['keyword']=='sinKeyword'].target.count())
```

```python
#hago un grafico de torta mostrando el porcentaje de tweets verdaderos sin keyword
plt.figure(figsize=(12, 8))
plt.pie(x=[prcSinKeyword,100-prcSinKeyword],labels=['Porcentaje de twits verdaderos sin keyword','Porcentaje de twits falsos sin keyword'],
        autopct="%.2f")
plt.suptitle("Porcentaje de veracidad twits verdaderos sin keyword",fontsize=20,fontweight="bold")
plt.show()
```

```python
prcConKeyword = (twit[twit['keyword']!='sinKeyword'].target.sum())*100/(twit[twit['keyword']!='sinKeyword'].target.count())
#hago un grafico de torta mostrando el porcentaje de tweets verdaderos sin keyword
plt.figure(figsize=(12, 8))
plt.pie(x=[prcConKeyword,100-prcConKeyword],labels=['Porcentaje de twits verdaderos con alguna keyword','Porcentaje de twits falsos con alguna keyword'],
        autopct="%.2f")
plt.suptitle("Porcentaje de veracidad de twits con alguna keyword",fontsize=20,fontweight="bold")
plt.show()
```

```python
#Analizo la cantidad de tweets por keyword
grouped = twit.groupby('keyword').agg({'text_long':['mean','max','min','std','count']})['text_long'].sort_values(by=['count'],ascending=False)
grouped.head(15)
```

```python
listaKey = grouped.reset_index().keyword.tolist()
twitsPorKey = grouped.reset_index()['count'].tolist()
```

```python
fig, ax = plt.subplots(figsize=(14,14))
ax.grid(True)
ax.set_ylabel('Cantidad de tweets')
ax.set_title('Distribucion de tweets segun por cada keyword',fontsize=20,fontweight="bold")
ax.set_xlabel('Keywords')
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) # labels along the bottom edge are off
plt.scatter(x=listaKey,y=twitsPorKey)
plt.savefig("image22.png")
```
